```python
#1. Find customers who haven't placed any orders
import pandas as pd
customers = pd.DataFrame({
    'customer_id1': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana']
})

orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'customer_id': [1, 2, 1]
})

merged_df = customers.merge(orders, left_on='customer_id1', right_on='customer_id', how='left')
merged_df[merged_df['order_id'].isna()]['name']
```

```python
#2. Find sudden temperature drops, use diff() function
df = pd.DataFrame({
    'date': pd.date_range(start='2024-01-01', periods=7),
    'temperature': [30, 28, 27, 15, 14, 25, 10]
})


df['temp_diff'] = df['temperature'].diff()
df[df['temp_diff'] < -10]
```

```python
days = np.arange(1, 8)
temperature_max = [30, 28, 27, 15, 14, 25, 10]
temperature_min = [20, 18, 17, 5, 4, 15, 0]

plt.figure(figsize=(10, 6))
plt.plot(days, temperature_max, label='Max Temperature', color='red', marker='o')
plt.plot(days, temperature_min, label='Min Temperature', color='blue', marker='o')
```

```python
plt.fill_between(days, temperature_max, temperature_min, color='pink', alpha=0.5)
```

```python
plt.fill_between(days, temperature_min, color='blue', alpha=0.3)
```

```python
days = np.arange(1, 8)
temperature_max = [30, 28, 27, 15, 14, 25, 10]
temperature_min = [20, 18, 17, 5, 4, 15, 0]

plt.figure(figsize=(10, 6))
plt.plot(days, temperature_max, label='Max Temperature', color='red', marker='o')
plt.plot(days, temperature_min, label='Min Temperature', color='blue', marker='o')
plt.fill_between(days, temperature_max, temperature_min, color='pink', alpha=0.5)
plt.fill_between(days, temperature_min, color='blue', alpha=0.3)
plt.title('Temperature Variation Over a Week')
plt.xlabel('Days of the Week')
plt.ylabel('Temperature (C)')
plt.legend(loc = 'upper right')
plt.show()
```
