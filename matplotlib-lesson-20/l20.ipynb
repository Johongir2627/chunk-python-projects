```python
#1. Find customers who haven't placed any orders
import pandas as pd
customers = pd.DataFrame({
    'customer_id1': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana']
})

orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'customer_id': [1, 2, 1]
})

merged = customers.merge(orders, left_on='customer_id1', right_on='customer_id', how='left')
merged[merged['order_id'].isna()]['name']
```

```python
#2. Find sudden temperature drops, use diff() function
df = pd.DataFrame({
    'date': pd.date_range(start='2024-01-01', periods=7),
    'temperature': [30, 28, 27, 15, 14, 25, 10]
})

df['change'] = df['temperature'].diff()
df[df['change'] < -10]
```

```python
days = np.arange(1, 8)
temperature_max = [28, 18, 26, 22, 19, 31, 15]
temperature_min = [18, 20, 15, 12, 29, 21, 17]

plt.figure(figsize=(10, 6))
plt.plot(days, temperature_max, color='red', marker='o', label = 'Max Temperature')
plt.plot(days, temperature_min, color='blue', marker='o', label = 'Min Temperature')
plt.fill_between(days, temperature_min, temperature_max, color='red', alpha=0.3)
plt.fill_between(days, temperature_min, color='blue', alpha=0.3)
plt.title('Weekly Temperature Change')
plt.xlabel('Day')
plt.ylabel('Temperature C')
plt.legend(loc='upper right')
plt.show()
```

```python
tips = pd.read_csv('tips.csv')
tips['per_person'] = tips['total_bill'] / tips['size']
tips['per_person'].plot.hist(bins=25)
```

```python
smokes = tips['smoker'].value_counts()
plt.pie(smokes, labels=smokes.index, autopct='%1.1f%%')
plt.show()
```

```python
smokes = tips['smoker'].value_counts()
plt.pie(smokes, labels=smokes.index, autopct=lambda p: f'{int(p*sum(smokes) / 100)}')
plt.show()
```
